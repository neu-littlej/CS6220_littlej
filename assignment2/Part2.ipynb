```python
%matplotlib inline
import pandas as pd

fileURL = 'pendigits.tra'
data = pd.read_csv(fileURL, header=None)

data['mean'] = data.mean(axis=1)
data['min'] = data.min(axis=1)
data['max'] = data.max(axis=1)
#data['mean'] = data.range(axis=1)
data['std'] = data.std(axis=1)
data['var'] = data.var(axis=1)
data['count'] = data.count(axis=1)

data = data.ix[:,'mean':]
```
