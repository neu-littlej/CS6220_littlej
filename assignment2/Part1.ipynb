```python
%matplotlib inline
import pandas as pd

fileURL = 'iris.data'
data = pd.read_csv(fileURL, names=['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width', 'Name'], header=None)

mean_flower = data.pivot_table(index='Name',  aggfunc='mean')
min_flower = data.pivot_table(index='Name',  aggfunc='min')
max_flower = data.pivot_table(index='Name',  aggfunc='max')
standev_flower = data.pivot_table(index='Name',  aggfunc='std')
#range_flower = data.pivot_table(index='Name',  aggfunc='range')
var_flower = data.pivot_table(index='Name',  aggfunc='var')
count_flower = data.pivot_table(index='Name',  aggfunc='count')

#perc_flower_25 = data.pivot_table(index='Name',  aggfunc='np.percentile(25)')
#perc_flower_50 = data.pivot_table(index='Name',  aggfunc='50')
#perc_flower_75 = data.pivot_table(index='Name',  aggfunc='75')
```
